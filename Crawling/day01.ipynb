```python
import requests # 웹 페이지 읽어오는 모듈 
from bs4 import BeautifulSoup as bs # 웹 페이지에서 긁어온 txt를 분석하기 쉽게 가공해주는 역할 
import pandas as pd
```

```python
url = "https://www.naver.com/"
txt = requests.get(url) # 웹 페이지 글자로 읽어오기 
html = bs(txt.text) # 글자를 html로 변환 
```

```python
##### Bugs 차트 읽어오기 #####
url = "https://music.bugs.co.kr/chart"
html = bs(requests.get(url).text)
```

```python
len(html.select('table.byChart > tbody > tr'))
```

```python
songs = html.select('table.byChart > tbody > tr')
print(songs)
```

```python
song = songs[0]
print(song)
```

```python
print(song.select('p.title > a'))
```

```python
print(song.select('p.title > a')[0].text)
```

```python
print(songs[1].select('p.title>a')[0].text)
```

```python
print(songs[0].select('p.artist > a')[0].text)
```

```python
song_data = []
rank = 1
for song in songs:
    title = song.select('p.title > a')[0].text
    singer = song.select('p.artist>a')[0].text
    song_data.append(['Bugs',rank,title,singer])
    rank += 1


df = pd.DataFrame(song_data, columns=['서비스','순위','타이틀','가수'])
df
```

```python
df.to_excel('bugs.xlsx',index=False)
```

```python
##### Melon 차트 읽어오기 #####
##### 막은거 뚫고 읽어오기 #####

url = "https://www.melon.com/chart/"
# html = bs(requests.get(url).text)
# print(html)

from selenium import webdriver

driver = webdriver.Chrome('chromedriver.exe') # 크롬에서 연 후 긁어오기 
driver.get(url)
txt = driver.page_source
html = bs(txt)
print(html)
```

```python
len(html.select('tbody > tr'))
```

```python
songs = html.select('tbody > tr')
print(songs[0])
```

```python
print(songs[0].select('div.rank01 > span > a')[0].text)
print(songs[0].select('div.rank02 > span > a')[0].text)
```

```python
song_data2 = []
rank = 1
for song in songs:
    title = song.select('div.rank01 > span > a')[0].text
    singer = song.select('div.rank02 > span > a')[0].text
    song_data2.append(['Melon',rank,title,singer])
    rank += 1


df = pd.DataFrame(song_data2, columns=['서비스','순위','타이틀','가수'])
df
```

```python
df.to_excel('melon.xlsx',index=False)
```

```python
driver = webdriver.Chrome('chromedriver.exe')
driver.get("https://www.genie.co.kr/chart/top200?ditc=D&ymd=20230227&hh=16&rtm=Y&pg=2")
html2 = bs(driver.page_source)
```

```python
songs50 = html1.select('tbody > tr')
songs100 = html2.select('tbody > tr')
song_data = []
rank = 1
for song in songs50:
    title = song.select('a.title')[0].text.strip()
    singer = song.select('a.artist')[0].text
    song_data.append(['Genie',rank,title,singer])
    rank += 1
rank = 51
for song in songs100:
    title = song.select('a.title')[0].text.strip()
    singer = song.select('a.artist')[0].text
    song_data.append(['Genie',rank,title,singer])
    rank += 1

df = pd.DataFrame(song_data, columns=['서비스','순위','타이틀','가수'])
df.to_excel('genie.xlsx',index=False)
```
