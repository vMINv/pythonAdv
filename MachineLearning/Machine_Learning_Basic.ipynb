```python
from sklearn.model_selection import train_test_split

X_data = df.loc[:,'sepal_length':'petal_width']
y_data = df.loc[:,'Target']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_data, y_data,
                                                    test_size=0.2,
                                                    # shuffle=True,
                                                    random_state=20)
```

```python
# KNN (K Nearest Neighbors) K 근접 이웃

# 예측 하려는 데이터 X를 좌표평면상에 위치시켜 기존 데이터와 가장 비슷한
# 이웃을 K개 찾는다. K값에 따라 모델의 예측력이 달라지므로 적절할 K를 찾아야한다. 

# 장점 : 수식에 관한 설명이 필요 없을 정도로 직관적이고 간단하다. 
# 단점 : 데이터가 많아질수록 상당히 느려진다. 또한 이상치에 취약하다. 

# 적용 : 주로 분류에 사용이 되며 작은 데이터 셋에 적합하다. (만개 정도도)
```

image.png

```python
# 모델 임포트
from sklearn.neighbors import KNeighborsClassifier

# 객체 생성
knn = KNeighborsClassifier(n_neighbors=21) 

# 학습
knn.fit(X_train, y_train)

# 예측
y_knn_pred = knn.predict(X_test)

# 평가
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_knn_pred)

# n_neighbors=3  0.9333333333333333
# n_neighbors=7  0.9666666666666667
# n_neighbors=11 1.0
# n_neighbors=21 0.9333333333333333

# shuffle=True,제거후  0.9333333333333333
```

```python
# 예측
knn.predict([[1.9,3.0,5.1,1.1]])
```

```python
# SVM Support Vector Machine
# 각 열들이 고유의 축을 갖는 벡터 공간을 이룬다고 가정한다. 
# 모든 데이터를 벡터 공간에 점으로 표시하고 각데이터가 속하는 목표 별로 군집
# 이룬다고 가정한다. 
# 각 군집까지의 거리를 최대한 멀리 유지하는 선을 찾는다. 따라서 각 군집을
# 서로 확연하게 구분 가능하게 한다. 
```

```python
# 모델 임포트
from sklearn.svm import SVC

# 객체 생성
svc = SVC(kernel='linear') 

# 학습
svc.fit(X_train, y_train)

# 예측
y_knn_pred = svc.predict(X_test)

# 평가
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_knn_pred)
```
