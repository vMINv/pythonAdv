```python
import matplotlib.pyplot as plt
```

```python
plt.plot([10, 20, 30, 40])
plt.show()
```

```python
plt.plot([1,2,3,4,],[12,43,25,15])
plt.show()
```

```python
plt.title('plotting')
plt.plot([10, 20, 30, 40])
plt.show()
```

```python
plt.title('legend')
plt.plot([10, 20, 30, 40], label='asc')
plt.plot([40,30,20,10], label='desc')
plt.legend()
plt.show()
```

```python
plt.title('color')
plt.plot([10, 20, 30, 40], color='yellow', label='yellow')
plt.plot([40,30,20,10], color='green', label='green')
plt.legend()
plt.show()
```

```python
plt.title('linestyle')
plt.plot([10, 20, 30, 40], color='r', linestyle='--', label='dashed')
plt.plot([40,30,20,10], color='g', ls=':', label='dotted')
plt.legend()
plt.show()
```

```python
plt.title('marker')
plt.plot([10, 20, 30, 40], 'r.', label='circle')
plt.plot([40,30,20,10], 'g^', label='triangle up')
plt.legend()
plt.show()
```

```python
plt.title('total')
plt.plot([10, 20, 30, 40], 'r.--')
plt.plot([40,30,20,10], 'g^:')
plt.show()
```
